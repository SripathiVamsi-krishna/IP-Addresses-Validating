```python
def valid(n):
    res = []
    try:
        ipaddress.ip_address(n)
        ip = n.split(".")
        res.append([int(x) for x in ip])
        res.append([format(int(x),'b') for x in ip])
        res.append([format(int(x),'o') for x in ip])
        res.append([format(int(x),'x') for x in ip])
        for i in range(len(res)):
            for j in range(len(res[i])):
                res[i][j]=str(res[i][j])
            res = ['.'.join(x) for x in res]
            print(res)
    except:
        res.append("Ip is not valid!")
    print(res)
```
